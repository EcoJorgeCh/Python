```python
#Parte1: Calculo de las condiciones de estudio
a1 = 0.0075
Ceo = 6  #Esto va a cambiar respecto al modelo matematico
import numpy as np
#If = np.array(list(range(100)))*100+100
np.random.seed(20) #Para que siempre seleccione el sorteo 4
e1 = np.random.uniform(0,99,100)
If = np.array(list(range(100,10100,100))) + e1
e2 = np.random.uniform(-17,17,100) #el últmo argumento es el número de datos
type(If)
Ce = Ceo + a1*If + e2
#Ce[50]=100  #Agregando una anomalia #Outlier
#print(len(If))
#print(len(Ce))
print("Si los ingresos familiares son igual a",If)
print("Las condiciones de estudio son igual a",Ce)
import matplotlib.pyplot as plt
plt.plot(If,Ce)
plt.title("Gráfico 1:\n Efeco del Ingreso Familiar \n en las Condiciones de Estudio")
plt.show()

plt.scatter(If,Ce)
plt.title("Gráfico 1A:\n Efeco del Ingreso Familiar \n en las Condiciones de Estudio")
plt.show()

Ce_hat = Ceo + a1*If #Parte determinística #Hat significa estimado
plt.scatter(If,Ce)
plt.plot(If,Ce_hat,color="red")
plt.title("Gráfico 1B:\n Efeco del Ingreso Familiar \n en las Condiciones de Estudio")
plt.xlabel("Ingreso Familiar")
plt.ylabel("Condiciones de estudios")
plt.show()
```

```python
#Parte 2: Creacion de la variable horas de lectura
e3 = np.random.uniform(-10,10,100)/100
Hl = np.array(list(range(0,1900,19)))/190 + e3
print(type(Hl))
print(len(Hl))
print("Las horas de lectura son igual a",Hl)
```

```python
#Parte 3: Elaboracion del Modelo final
Co = 3.1
b1 = 0.5
e4 = np.random.uniform(-5,5,100)
Cc = Co*Hl**b1*Ce**(1-b1) + e4
print(len(Cc))
print("Entonces la comprensión  del curso es igual a",Cc)
plt.plot(Hl,Cc)
plt.title("Gráfico 2:\n Impacto de las horas de \n estudio en la Comprension del Curso")
plt.show()

plt.scatter(Hl,Cc)
plt.title("Gráfico 2A:\n Impacto de las horas de \n estudio en la Comprension del Curso")
plt.show()

Cc_hat = Co*Hl**b1*Ce_hat**(1-b1)
plt.scatter(Hl,Cc)
plt.plot(Hl,Cc_hat,color="red")
plt.title("Gráfico 2B:\n Impacto de las horas de \n estudio en la Comprension del Curso")
plt.xlabel("Horas de Lectura")
plt.ylabel("Comprensión del curso")
plt.show()
```

```python
#Parte 4: Gráficos del Modelo
import matplotlib.pyplot as plt
plt.plot(If,Ce)
plt.title("Gráfico 1:\n Efeco del Ingreso Familiar \n en las Condiciones de Estudio")
plt.show()
plt.plot(Hl,Cc)
plt.title("Gráfico 2:\n Impacto de las horas de \n estudio en la Comprension del Curso")
plt.show()
plt.plot(Ce,Cc)
plt.title("Gráfico 3:\n Impacto de las Condiciones de \n Estudio en la Comprension del Curso")
plt.show()
plt.plot(Hl,Ce)
plt.title("Gráfico 4:\n Relacion entre las \n Horas de Lectura y \n las Condiciones de Estudio")
plt.show()
plt.plot(Cc)
plt.title("Gráfico 5:\n Evolución de Comprension Lectora")
plt.show()
```
