```python
import numpy as np
# Y=A*K**a1*L**(1-a1)
# y=A*k**a1
a1 = 0.4
A = 2.5
k = np.array(list(range(0,100,1))) 
y = A*k**a1
import matplotlib.pyplot as plt
plt.plot(k,y)
```

```python
#S=Y-C
#S=sY
#S/L = s*Y/L = s*y = S_L
s=0.5
S_L = s*y
plt.plot(k,S_L)
```

```python
plt.plot(k,y)
plt.plot(k,S_L)
```

```python
#D=delta*k
d = 0.106
D = d*k
plt.plot(k,y)
plt.plot(k,S_L)
plt.plot(k,D)
plt.title("Grfico 3:\n Modelo Solow-Swan")
plt.legend(['y', 'S_L', 'D'])
plt.xlabel("Capital Per-Capita")
```
